```python
import pandas as pd

df = pd.read_csv("netflix_titles.csv")
print(df.head())
```

```python
print(f"Original datasett: {len(df)}")

df = df.dropna()
df = df.drop_duplicates()
print(f"Clean dataset: {len(df)}")
```

```python
movies = df[df['type']=='Movie']
longest = movies.loc[movies['duration'].str.replace(' min','').astype(int).idxmax()]
shortest = movies.loc[movies['duration'].str.replace(' min','').astype(int).idxmin()]
print(f"Longest Movie: {longest['title']} ({longest['duration']})")
print(f"Shortest Movie: {shortest['title']} ({shortest['duration']})")
```

```python
sorted = df.sort_values(by='release_year', ascending=True)
# Reset index og kast gammel index
df_sorted = sorted.reset_index(drop=True)
df_sorted.to_excel("Netflix_history_sorted.xlsx", index=False)
print(df_sorted.head())
```

```python
#Antall filmer/serier per direktor
director_count = df.groupby("director").size().sort_values(ascending=False)
print(director_count)
```
